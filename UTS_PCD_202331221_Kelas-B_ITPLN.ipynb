# Import Library

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Membaca gambar

```python
color_image = cv2.imread('Warna.jpg')
plt.imshow(color_image)
```

# ukuran gambar

```python
color_image.shape
```

# mengubah dari bgr menjadi rgb

```python
brg = cv2.cvtColor(color_image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(brg)
```

# Mendeteksi warna citra

```python
plt.subplot(2, 2, 1)
plt.imshow(brg)
plt.title('Original Image')

plt.subplot(2, 2, 2)
plt.imshow(cv2.cvtColor(brg, cv2.COLOR_BGR2RGB)[:, :, 2], cmap="gray")
plt.title('Red Channel')

plt.subplot(2, 2, 3)
plt.imshow(cv2.cvtColor(brg, cv2.COLOR_BGR2RGB)[:, :, 1], cmap="gray")
plt.title('Green Channel')

plt.subplot(2, 2, 4)
plt.imshow(cv2.cvtColor(brg, cv2.COLOR_BGR2RGB)[:, :, 0], cmap="gray")
plt.title('Blue Channel')

plt.tight_layout(pad=3.0)  # Tambahkan ini untuk memberi jarak antar subplot
plt.show()
```

# Mendeteksi warna citra dan menampilkan histogram

```python
#Merah
merah=cv2.cvtColor(brg, cv2.COLOR_BGR2RGB)[:,:,2] 
fig, axs = plt.subplots(3,2, figsize = (15,5))
hist = cv2.calcHist([merah],[0],None,[256],[0,256])
axs[0,0].imshow(merah, cmap='gray')
axs[0,1].plot(hist)

#Hijau
hijau=cv2.cvtColor(brg, cv2.COLOR_BGR2RGB)[:,:,1] 
hist = cv2.calcHist([hijau],[0],None,[256],[0,256])
axs[1,0].imshow(hijau, cmap='gray')
axs[1,1].plot(hist)

#Biru
biru=cv2.cvtColor(brg, cv2.COLOR_BGR2RGB)[:,:,0] 
hist = cv2.calcHist([biru],[0],None,[256],[0,256])
axs[2,0].imshow(biru, cmap='gray')
axs[2,1].plot(hist)
plt.show()
```
